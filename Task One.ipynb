# Linear regression with sinusoide fuction

```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt 
from datetime import date, timedelta
```

```python
df = pd.read_csv('Nat__Gas.csv', parse_dates=['Dates'])
prices = df['Prices'].values
dates = df['Dates'].values
```

### Plotting prices against dates

```python
fig, ax = plt.subplots()
ax.plot_date(dates, prices, '-')
ax.set_xlabel('Dates')
ax.set_ylabel('Prices')
ax.set_title('Natural Gas Prices')
ax.tick_params(axis='x', rotation=45)
```

It is noticeable from the plot there is a "linear" relationship around the year as a natural frequency that trend upwards. 

Hence, it can be used a linear regression to estimate this trend and combined with sin function to take account of the variation in each year.

### Converting dates to days from the start

```python
start_date = date(2020,10,31)
end_date = date(2024,9,30)
months = []
year = start_date.year
month = start_date.month + 1
while True:
    current = date(year, month, 1) + timedelta(days= -1)
    months.append(current)
    if current.month == end_date.month and current.year == end_date.year:
        break
    else:
        month = ((month + 1) % 12) or 12
        if month == 1:
            year += 1
            
days_from_start = [(day - start_date).days for day in months]
```

### Simple regression 

```python
def simple_regression(x, y):
    x_bar = np.mean(x)
    y_bar = np.mean(y)
    slope = np.sum((x - x_bar) * (y - y_bar)) / np.sum((x - x_bar) **2)
    intercept = y_bar - slope * x_bar
    return slope, intercept    
```

```python
time = np.array(days_from_start)
slope, intercept = simple_regression(time, prices)
```

```python
#plot linear trend
plt.plot(time, prices)
plt.plot(time, time * slope + intercept)
plt.xlabel('Days from start date')
plt.ylabel('Prices')
plt.title('Linear Trend Visualisation')
plt.show()
```

It can be capture that there is linear trend. 
However, it can be gather from the plot that the frequency of price movement is about a year (12 months). In fact, natural gas mostly used in winter and less in summer. 

Hence, looking to the graph it can be noticed a resemblance to a sinusoide function( y = amp * Sin(k*t + z) ). 
It can be rewritten y = Cos(z) * Sin(kt) + Sin(z) * Cos(kt), in this way it can be solved by using a bilinear_regression with no intercept.

### Bilinear regression 

```python
sin_prices = prices - (slope * time + intercept)
sin_time = np.sin(time * 2 * np.pi / (365))
cos_time = np.cos(time * 2 * np.pi / (365))
```

```python
def bilinear_regression(y, x_1, x_2):
    slope_1 = np.sum(y * x_1) / np.sum(x_1 ** 2)
    slope_2 = np.sum(y * x_2) / np.sum(x_2 ** 2)
    return(slope_1, slope_2)
```

```python
slope_1, slope_2 = bilinear_regression(sin_prices, sin_time, cos_time)
```

```python
# recovering amplitude (~ height of the sinusoide) and phase shift 
#Amp = sqrt of (slope1 ^ 2 + slope2 ^ 2)
#P. shift(z) = arctan of (slope2/slope1)
amp = np.sqrt(slope_1 ** 2 + slope_2 ** 2)
shift = np.arctan2(slope_2, slope_1)
```

```python
#plot estimate to the dataset
plt.plot(time, amp * np.sin(time * 2 * np.pi / (365) + shift))
plt.plot(time, sin_prices)
plt.title('Sinusoide Estimate Visualisation')
```

```python
# interporlation/extrapolation  fuction 
def extrapolation(date):
    days = (date - pd.Timestamp(start_date)).days
    if days in days_from_start: # exact match found
        return prices[days_from_start.index(days)]
    else: # use sin/cos model to interpolate => extrapolate
        return amp * np.sin(days * 2 * np.pi / (365) + shift) 
    + days *slope + intercept
        
```

```python
# range of continuos days by converting the dates 
continuos_dates = pd.date_range(start = pd.Timestamp(start_date),
                                end = pd.Timestamp(end_date),
                               freq = 'D')
```

```python
#plot estimate using the interpolation/extrapolation func
plt.plot(continuos_dates, [extrapolation(date) for date in continuos_dates],
        label = 'Sinusoide Estimate with Interpolation')
```

### Fitting the pries with sinusoide curve

```python
x = np.array(days_from_start)
y = np.array(prices)
fit_amp = np.sqrt(slope_1 ** 2 + slope_2 ** 2)
fit_shift = np.arctan2(slope_2, slope_1)

fit_slope, fit_intercept = simple_regression(x, y - fit_amp 
                                             * np.sin(x * 2 * np.pi / (365)
                                                     + fit_shift))
plt.plot(dates, y, 'o', label='Monthly Input Prices')

y_fit =  fit_amplitude * np.sin((continuos_dates - pd.Timestamp(start_date)).days \
                                * 2 * np.pi / 365 + fit_shift) \
+ (continuos_dates - pd.Timestamp(start_date)).days * fit_slope + fit_intercept
 
plt.plot(continuos_dates, y_fit, label='Fitting to Sin Curve')

plt.xlabel('Dates')
plt.ylabel('Prices')
plt.title('Natural Gas Prices')
plt.legend()
plt.show()
```
